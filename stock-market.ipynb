```python
# Forecasting Model
```

```python
X = icici_df[['MA20', 'MA50', 'Returns', 'Volatility']]
y = icici_df['Target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    shuffle=False
)
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

scaler = MinMaxScaler(feature_range=(0,1))

scaled_data = scaler.fit_transform(icici_df[['Close']])
```

```python
X_lstm = []
y_lstm = []

for i in range(60, len(scaled_data)):
    X_lstm.append(scaled_data[i-60:i, 0])
    y_lstm.append(scaled_data[i, 0])

X_lstm = np.array(X_lstm)
y_lstm = np.array(y_lstm)
```

```python
X_lstm = np.reshape(
    X_lstm,
    (X_lstm.shape[0], X_lstm.shape[1], 1)
)
```

```python
split = int(len(X_lstm) * 0.8)

X_train_lstm = X_lstm[:split]
X_test_lstm = X_lstm[split:]

y_train_lstm = y_lstm[:split]
y_test_lstm = y_lstm[split:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model_lstm = Sequential()

model_lstm.add(LSTM(
    units=50,
    return_sequences=True,
    input_shape=(X_train_lstm.shape[1], 1)
))

model_lstm.add(LSTM(
    units=50,
    return_sequences=False
))

model_lstm.add(Dense(units=25))
model_lstm.add(Dense(units=1))
```

```python
model_lstm.compile(
    optimizer='adam',
    loss='mean_squared_error'
)
```

```python
model_lstm.fit(
    X_train_lstm,
    y_train_lstm,
    batch_size=32,
    epochs=10
)
```

```python
lstm_predictions = model_lstm.predict(X_test_lstm)
```

```python
lstm_predictions = scaler.inverse_transform(lstm_predictions)

y_test_actual = scaler.inverse_transform(
    y_test_lstm.reshape(-1,1)
)
```

```python
from sklearn.metrics import mean_absolute_error

lstm_mae = mean_absolute_error(
    y_test_actual,
    lstm_predictions
)

print("LSTM MAE:", lstm_mae)
```

```python
plt.figure(figsize=(14,7))

plt.plot(y_test_actual, label='Actual Price')
plt.plot(lstm_predictions, label='LSTM Predicted Price')

plt.legend()
plt.title('LSTM Actual vs Predicted')
plt.show()
```

```python
def forecast_next_days(model, last_sequence, days=7):
    future_predictions = []
    current_sequence = last_sequence.copy()

    for _ in range(days):
        pred = model.predict(current_sequence.reshape(1, -1))
        future_predictions.append(pred[0])
    return future_predictions
```
